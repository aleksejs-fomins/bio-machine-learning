$$\dot{x}(t + 1) = \Theta \biggl(Mx(t) + I(t) - T\biggr)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nNeuron = 100

def simulate(M, nT, T):
    x = np.zeros((nT, nNeuron))
    x[0] = np.random.uniform(0, 1, nNeuron)
    
    for i in range(nT-1):
        I = np.random.uniform(0, 1.3, nNeuron) #+ 0.1 * np.sin(2 * np.pi * i / 10)
        x[i+1] = (M.dot(x[i]) + I > T).astype(int)

    xMag = np.sum(x, axis=1)
        
    plt.figure()
    plt.plot(xMag)
    plt.show()
    
    return x
    
M = np.random.uniform(0, 0.018, (nNeuron, nNeuron))    
x = simulate(M, 10000, 1)
```
